# Les listes
Les listes sont des structures de données qui permettent de stocker plusieurs valeurs dans un ordre précis.\
Plus concrètement :

```python
une_liste = ["a",3,True]
```

Pour créer une liste, on met donc plusieurs valeurs entre crochets, séparées par des virgule :\
[ valeur_a, valeur_b, ... ]

Chaque valeur se trouve à un certain **INDICE** : c'est son n° dans la liste : 
- valeur_a est à l'indice 0
- valeur_b est à l'indice 1
- etc.

Pour afficher une valeur on va utiliser le nom de la liste, des crochets, et le n° d'indice :

```python
une_liste = ["a",3,True]
print( une_liste[0] ) # affiche "a"
print( une_liste[1] ) # affiche 3
print( une_liste[2] ) # affiche True
```

```python
On peut aussi afficher tous les éléments :
```

```python
une_liste = ["a",3,True]
print(une_liste)
```

Ou les deux premiers :

```python
une_liste = ["a",3,True]
print(une_liste[:2])
```

Ou les deux dernier :

```python
une_liste = ["a",3,True]
print(une_liste[-2:])
```

Ceux du milieu : 

```python
une_plus_grande_liste = ["a",3,True,"5", "6", 3]
print(une_plus_grande_liste[2:5]) # l'indice 5 est exclu ! ça correspond en math à [2:5[ 
```

## Modifier un élément
Comme pour modifier une variable !
il suffit juste de préciser quel indice on veut changer ! 

```python
liste = ["fraises","pommes"]
print(liste)
liste[1] = "asperges"
print(liste)
```

## Parcourir une liste
On va bien évidemment utiliser *for* : 

```python
fruits = ["fraises","cerises","framboises","mirabelles"]
print("cherchez l'intrus!")
for fruit in fruits :
    print("les", fruit, "sont un fruit de printemps")
```

## Ajouter un élément dans une liste
On va utiliser les fonctions *append(x)* (insère un élément à la fin de la liste) ou *insert(i,x)* (insère un élément à la position indiquée) : 

```python
animals = ["chat","chien"]
animals.append("perroquet")
animals.insert(0,"ara")
print(animals)
```

## Supprimer un élément dans une liste
On va utiliser les fonctions *remove(x)* (supprime de la liste le premier élément dont la valeur est égale à x) ou *pop()* (supprime le dernier élément de la liste et le retourne) ou *pop(x)* (supprime l'élément à l'indice *x*):

```python
cats = ["maincoon","angora","persan","siamois","européen"]
cats.remove("siamois")
triton = cats.pop(0)
ajax = cats.pop()
print(cats)
print(ajax)
```
