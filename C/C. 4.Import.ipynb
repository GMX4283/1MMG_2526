# import
Python permet d’importer des modules (des fonctionnalités supplémentaires) dans votre code en utilisant le mot clé import :

```python
import math
import random

print(random.randint(1,10))
print( math.sqrt(9))
```

Un module peut-être un autre de vos fichiers .py ou une bibliothèque contenant des 
fonctions supplémentaires : math, random, sys… (s’aider de Google pour trouver la 
bonne!)
