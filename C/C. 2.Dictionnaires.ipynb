# Les Dictionnaires
Ce sont des listes qui utilisent des clés plutôt que des indices :

```python
ma_voiture = {"brand":"dacia","year":1995,"color":"blue"}
print(ma_voiture)
```

Pour créer un dictionnaire, on utilise les accolades { }.\
Puis, pour chaque élément on va préciser la clé et la valeur :\
*dico_name = {key1:value1, key2:value2, ...}*

Pour afficher une valeur on va utiliser la clé :

```python
print(ma_voiture["color"])
```

## Modifier un élément
il suffit juste de préciser la clé que l'on veut modifier : 

```python
ma_voiture["brand"] = "Aston Martin"
print(ma_voiture)
```

## Ajouter un élément dans un dictionnaire
On va écrire le même code que pour modifier la liste !\
Si Python ne trouve pas la clé, il la crée :

```python
ma_voiture["HP"] = "365ch"
print(ma_voiture)
```

## Supprimer un élément dans un dictionnaire
On va utiliser la fonction *pop(key)* : 

```python
ma_voiture.pop("color")
print(ma_voiture)
```

*clear()* permet de tout supprimer :

```python
ma_voiture.clear()
print(ma_voiture)
```
