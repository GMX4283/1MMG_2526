# Fonctions
Rappel : C’est une instruction ou une expression, identifiée par un nom, contenant une séquence : cela permet de créer une séquence réutilisable. 

## Définition 
Pour créer une fonction, on doit la « définir » :\
On donne un nom à la séquence qui composera notre fonction. 

## Appel 
Pour exécuter une fonction, il suffit d’écrire son nom dans une autre instruction :\
On dit qu’on appelle la fonction.

```python
# on definit (avec le mot clé def) la fonction appelée affiche_un_truc
def affiche_un_truc() : 
    print("un truc")

# on appelle affiche_un_truc
affiche_un_truc()
# une deuxième fois
affiche_un_truc()
```

## Argument 
Quand on définit notre fonction, on peut préciser si elle aura des arguments :\
Ce seront des variables déclarées et affectées lors de l’exécution de la fonction, contenant les valeurs passées en paramètre. 
## Paramètres  
Si la fonction a des arguments on doit donner la valeur de chaque argument lors de l’appel à la fonction :\
On dit qu’on « passe des paramètres » à la fonction

```python
# on definitla fonction 'affiche_ce_truc' avec un argument appelé 'truc_à_afficher'
def affiche_ce_truc(truc_à_afficher) : 
    print(truc_à_afficher)

# on appelle 'affiche_ce_truc' en passant des paramètres qui deviendront les 'truc_à_afficher'
affiche_ce_truc("Hey") 
# une deuxième fois
affiche_ce_truc("Hey")
```

## Valeur de retour
Pour éviter de perdre la valeur calculée dans la fonction, on peut la retourner :\
A la fin de l’exécution de la fonction, on peut demander à l’ordinateur de retourner une ou plusieurs valeurs à la séquence parente. 

```python
def cest_pas_faut(mot_pas_compris) : 
    texte = "C'est " + mot_pas_compris + " que vous comprenez pas ?"
    return texte

print(cest_pas_faut("côtelettes"))
print(cest_pas_faut("la ref à Kaamelott"))
```
