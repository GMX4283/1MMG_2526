# Variables
Une variable permet de stocker une valeur dans la mémoire de notre ordinateur. \
On peut la voir comme une boite dans laquelle on met une information qu'on utilisera plus tard. \
Pour créer une variable, il faut la nommer et affecter ("mettre") une valeur dedans. 
> IMPORTANT :
> 1) l'opérateur d'affectation est le signe = \
Ce n'est pas pour indiquer l'égalité comme en mathématiques !!
> 2) la variable dans laquelle on veut mettre une valeur est TOUJOURS à gauche du =

La syntaxe pour créer et affecter une variable est la suivante :

```python
ma_variable = 12
```

Ici, j'ai créé une variable appelée ma_variable et mis la valeur 12 dedans.\
Si je réécris ma_variable, Python va aller chercher la valeur en mémoire :

```python
print(ma_variable)
```

print() permet d'afficher un message dans la CONSOLE.\
Je demande ici d'afficher ma_variable, plus précisément son contenu.\
Il m'affiche donc logiquement 12!\
L'idée est maintenant de tous le temps utiliser des variables pour stocker les valeurs utiles à notre programme.\
Par exemple, pour calculer mon âge, ne pas faire :

```python
print(2025 - 1983 )
```

Mais plutôt essayer de faire :

```python
annee_courante = 2025
annee_naissance = 1983
age = annee_courante - annee_naissance
print(age)
```

Pourquoi ? Parce que on fait de la PROGRAMMATION, c'est à dire on crée des programmes !\
Un programme est "une suite d'actions que l'on se propose d'accomplir pour arriver à un résultat" (Robert).\
En informatique, pour programmer on va donc devoir déduire quelles actions on va devoir faire pour arriver à ce résultat.\
Ainsi, si on veut faire un programme pour trouver l'âge (le résulat attendu) en fonction de l'année de naissance et l'année courante (les données d'entrée), on ne peut pas savoir si la bonne réponse est 17, 42, 33 ou 6 : cela va changer pour chaque personne !\
Par contre on va devoir créer un programme qui donnera toujours la bonne réponse en fonction de l'année courante et de la date de naissance...\
On avait déjà vu ce programme :

```python
annee_courante = 2025
annee_naissance = 1983
age = annee_courante - annee_naissance
print(age)
```

Le programme ci-dessous, plus complexe, va vous demander votre année de naissance et calculer votre âge en fonction de l'année courante :

```python
from datetime import datetime
annee_courante= current_year = datetime.now().year
annee_naissance = int(input("quel est ton année de naissance ?"))
age = annee_courante - annee_naissance
print("tu auras",age,"ans cette année")
```

Et celui-ci est encore plus complexe, mais plus précis !

```python
from datetime import datetime
import math
while True:
    try:
        date_naissance_str = input("Donne ta date de naissance au format JJ.MM.AAAA : ").strip()
        date_naissance = datetime.strptime(date_naissance_str, "%d.%m.%Y").date()        
        break
    except ValueError:
        print("Format de date invalide. le format est JJ.MM.AAAA. Exemple : 05.05.1983")
aujourdhui = datetime.now().date()
date_anni = datetime(aujourdhui.year, date_naissance.month, date_naissance.day).date()
age = math.floor((aujourdhui - date_naissance).days/365)
jour_anni = (date_anni - aujourdhui).days  
print("tu as", age ,"ans.")
if jour_anni > 0 :
    print("ton anniversaire est dans",jour_anni,"jours!")
else :
    if jour_anni < 0 :
        print("ton anniversaire est dans",jour_anni+365,"jours!")
    else : 
        print("Joyeux anniversaire !")
```
