# Les types en Python

**Une variable ne peuvent contenir qu'un seul type de données à la fois.**

Les 5 types principaux sont :

| Type | Nom en python | Exemple |
| --- | ----------- | ----- | 
| Nombre entier (integer) | int | 5 |
| Nombres réels (float) | float | 5.4321 |
| Texte/Chaines de caractères (string) | str | "Hello world" |
| Booléen | bool | False |

```python
variableNombreEntier = 42
variableNombreReel = 3.14 # !!! le séparateur de décimale est le . en Python !
variableTexte = "J'ai mangé trop de soupe"
variableBooleen = True

# affiche le type de chaque variable
print("variableNombreEntier est de type", type(variableNombreEntier))
print("variableNombreReel est de type", type(variableNombreReel))
print("variableTexte est de type", type(variableTexte)) 
print("variableBooleen est de type", type(variableBooleen)) 
```

On ne peut pas mélanger les types quand on fait des opérations, sauf int et float :

```python
a = 2
b = 5.2
print(a*b) # OK

c = 2 
d = "4" # ce n'est pas un int mais une string !
print(c+d) # ERREUR !
```

Il faut convertir une des deux variables :

```python
d = int(d) #  1) int(d) -> on convertit d en int  2) d = int(d) -> on remet d convertit dans d
print(c+d)
```

# Nommer sa variable

En Python, le nom d’une variable peut contenir n'importe quelle suite de lettres (sans accent de préférence), de chiffres, ainsi le caractère ' _ '.

Par contre, dans le nom, on ne peut pas :
- mettre d'espace
- commencer par un chiffre
- utiliser un "mot-clé" du langage (if, else, is, break...)

Exemples OK:

```python
Chiffre1 = 1
mon_chat = "mallow"
Vré = True 
```

Exemple KO :

```python
1Chiffre = 1
mon chat = "mallow"
True = True 
```
