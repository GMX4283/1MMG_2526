# Instructions conditionnelles

Si on veut exécuter des instructions que dans certains cas, il faut utilisier un **test** qui va nous dire si la condition est remplie ou non.\
Pour tester la condition, on utilisera les opérateurs de comparaison.\
Et on dira alors au programme :\
**Si** *condition == VRAI* :
> *instructions* si c'est vrai remplie

**Sinon** :
> *instructions* si c'est faux

En Python on utilise  *if, else*:

```python
# seulement avec "si"
if 3 < 4 :
    print("la proposition 3 < 4 est vraie !")
```

```python
#si / sinon
if 4 < 3 :
    print("la proposition 3 < 4 est vraie !")
else :
    print("PAS VRAI! on a inversé 3 et 4")
```

L'important est que le résulat de l'assertion soit VRAI, même si la proposition est fausse. Exemple :

```python
#si / sinon
ma_proposition = False
if ma_proposition == False :
    print("c'est vrai que ma_proposition est égale à False ")
else :
    print("c'est faux que ma_proposition est égale à False")
```

#### Indentation

L'indentation est importante !!!!\
C'est l'espace devant la ligne de code.\
Elle permet de créer des "blocs" : des instructions qui s'exécuteront que dans certaines conditions.\
Le logiciel crée souvent la bonne indentation tout seul.\
Sinon on doit rajouter des espaces, ou MIEUX, appuyer sur la touche TAB !\
Une mauvaise indentation peut radicalement changer la signification du code.\
C'est mieux comme ça... :

```python
perdu = True
print("Tu es un...")
if perdu == True : 
    print("Loser...")
else : 
    print("Winner !")
print("Continue comme ça !")
print("Bye")
```

#### Imbriquer les IF et les ELSE

On peut combiner les IF et les ELSE pour obtenir du code plus complexe.\
C'est utile quand on a plus d'un cas à tester.\
Exemples :

```python
nombre_plats = 1
if nombre_plats == 1:
    print("juste le plat")
else : 
    if nombre_plats == 2:
        print("entrée ou dessert ?")
    else :
        if nombre_plats == 3:
            print("la totale !")
        else : 
            print("le serveur n'a pas compris")
```

On peut aussi utiliser ELIF pour que ça soit plus lisible :

```python
nombre_plats = 1
if nombre_plats == 1:
    print("juste le plat")
elif nombre_plats == 2:
    print("entrée ou dessert ?")
elif nombre_plats == 3:
    print("la totale !")
else : 
    print("le serveur n'a pas compris")
```

Au contraire, parfois on veut tester plusieurs conditions, les unes après les autres !

```python
age = 18

if age > 17 : 
    print("tu es majeur.e")

if age < 25 : 
    print("tu es né.e ce millénaire")

if age < 18 : 
    print("tu es né.e après l'iPhone")

if age > 40 : 
    print("ta date de naissance est plus proche de la fin de la 2ème guerre mondiale que l'année en cours")    
```

#### Opérateurs logiques

Pour rappel, on peut combiner des conditions entre elles avec des opérateurs logiques.\
Ce qui permet de faire des codes de plus plus complexes (et intelligents)!

```python
jaime_Toto = True
jaime_Titi = False

if jaime_Toto == True and jaime_Titi == True :
    print("I'm in love ! Mais j'ai un problème ^^ ")
else:
    if jaime_Toto == True or jaime_Titi == True :
        print("I'm in love !")
    else : 
        print("I'm not in love !")
```
