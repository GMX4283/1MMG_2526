# Les opérateurs

En Python, pour effectuer des opérations arithmétiques ont peut utiliser les opérateurs suivants :

|opérateur|nom|exemple|
|----|----|----|
|+|addition| 3 + 3 |
|-|soustraction| 1 - 2 |
|*|multiplication| 5 * 8 |
|/|division| 8 / 4 |
|**|puissance| 3**2 |
|//|division entière| 9//2 |
|%|modulo|5%2|

Exécute tous les exemples ci-dessous :

```python
3 + 3
```

```python
4 * 5 
```

```python
4**5
```

```python
5 // 2
```

```python
5 % 2
```

```python
4 * (5 + 3)
```

#### Opérateurs et strings
Les opérateurs peuvent aussi être utilisés avec des chaînes de caractère (string).
Par exemple pour coller 2 chaines ensembles : +

```python
"toto "+ "et " + "titi" 
```

C'est ce qu'on appelle une **concaténation**!\
On peut aussi utiliser le signe multiplier pour écrire plusieurs fois la même chose :

```python
"ah" * 3
```

# Opérateurs de comparaison

Les opérateurs de comparaison sont :

|Comparaison | Opérateur |
| ------| ----------|
|est égal | == |
|n'est pas égal ≠ | != |
|plus petit que | < |
|plus grand que | > |
|plus petit ou égal ≤ | <= |
|plus grand ou égal ≥ | >= |

On va comparer la valeur à gauche avec la valeur à droite. Exemple :\
`a = 3`\
`b = 5`\
`a > b`

**En fonction de l'opérateur, l'ordinateur va nous dire si le résulat de l'opération logique :** 
- est VRAI, en Python *True*
- est FAUX, en Python *False*

`a > b` va donc avoir pour résultat `False`

```python
a = 3
b = 5
print(a > b)
```

`a > b` est une expression. Comme toutes les expressions, elle est remplacée par son résultat lors de l'exécution du programme.\
C'est pour ça que le programme affiche True ou False.\
**ATTENTION! :** *==* est l'opérateur de comparaison de l'égalité ! Ce n'est pas *=*  

```python
a = 5
b = 5
print(a == b)
```

Autres exemples :

```python
print(4 == 5)
print(4 == 4) 
print(6 >= 0)
print(-1 < -5)
print(False == False) 
print("toto" != "titi")
print("titi" != "titi")
```

# Opérateurs logiques

On peut combiner les opérations logiques entre elles avec des opérateurs logiques.\
Il y en a 3 : *and, or* et *not*

```python
print(4 == 5 and 4 == 4) 
print(4 < 5 and 4 == 4) 
print(4 == 5 or 4 == 4) 
```

```python
nombre = 10
if nombre >= 0 and nombre <= 9:
    print("c'est un chiffre")
```

Le *or* renvoie *True* si une des deux opération est *True*

```python
jaime_Toto = True
jaime_Titi = False
if jaime_Toto == True or jaime_Titi == True :
    print("I'm in love !")
```

Le *not* est rarement utilisé :

```python
print(not 4 == 5)
print(not 4 == 4) 
```
