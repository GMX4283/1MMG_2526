# print() = AFFICHER
Cette fonction permet d'afficher ce qu'on met dans les parenthèse. Dans ces parenthèses on peut mettre : 
- une ou des valeurs
- une ou des variables
- des opérations, des fonctions... (dans ce cas le résultat s'affiche)
- mélanger tout ça en séparant par une virgule !

```python
print(98)
print(7+8)
a = 5
b = 6
print(a+b)
print(a+b+9)
print("toto")
le_chat = "ajax"
print(le_chat)
print("mon chat s'appelle", le_chat, "(et pas le_chat, c'est le nom de la variable!)")
print(5*"miaou ")
print(3,"petits chats")
from datetime import datetime
print("il est",datetime.now().time())
```

# input() = DEMANDER

- il faut TOUJOURS stocker la valeur dans une variable !

Quand l'ordinateur lit input dans votre code, il arrête le programme et attend que vous répondiez.\
N'oubliez pas d'appyuer sur la touche Entrée !

```python
jai_dit = input()
print("j'ai dit : ", jai_dit)
```
